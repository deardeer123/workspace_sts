```python
# 한국전기안전공사_연도별 감전사고인명피해현황 시간별_20221231
df = pd.read_csv("한국전기안전공사_연도별 감전사고인명피해현황 시간별_20221231.csv", encoding='CP949')

# 위 데이터프레임을 다른 데이터프레임 처럼 
# 인덱스를 연도로 바꾸기 위해 발생시간이 열(속성)으로 보내고
# 연도 를 행(인덱스)로 만들 필요가 있음
df
```

```python
#인덱스 제거 및 재지정
df.set_index('발생시간', inplace=True)

#인덱스를 재지정한 데이터프레임 확인
df
```

```python
# 행과 열을 바꾸는 작업을 수행해야함 -> transpose() 사용 
df = df.transpose()
df
```

```python
# 인덱스 이름이 사라지질 않아서 딕셔너리 형식으로 바꾼 후 다시 데이터프레임으로 변환
new_dict = df.to_dict()

df = pd.DataFrame(new_dict)
```

```python
#데이터 베이스에 Insert 하기 위한 함수
def insertData(dict, conn):
    cursor = conn.cursor(pymysql.cursors.DictCursor)
    columns = list(dict.keys())
    
    sql = "INSERT INTO YEAR_BY_TIME"
    sql = sql + " VALUES "

    index = columns[0]
    for i in (range(len(dict[index]))):
    
        for idx,column in enumerate(columns) :
        
            if(idx == 0):
                sql = sql + "("
            sql = sql + f"{dict[column][i]}"

            if(idx+1 == len(columns)):
                sql = sql + ")"
            else:
                sql = sql + ","
        if(i+1 != len(dict[index])):
            sql = sql + ',' 
    
    print(sql)
    cursor.execute(sql)
    conn.commit()
        
```
