```python
#db 연결하는 함수
def getConnection():
    # 1. 파이썬에 데이터베이스 연결하는 메소드
    conn = pymysql.connect(
        #데이터베이스 IP
        host='localhost',
        #데이터베이스 사용자 이름
        user='team2_user',
        #데이터베이스 사용자 비밀번호
        password='team2',
        #접속하고자 하는 데이터베이스
        db='team2' ,
        #문자 인코딩 (데이터베이스로부터 데이터를 받아올때 한글이 깨지는것을 막기 위함)
        charset='utf8')
    
    #connect객체 리턴
    return conn
```

```python

# 테이블 생성을 위한 함수
# 매개변수로 create_sql_list(테이블을 생성하기위한 쿼리문이 담긴 리스트),
# connection 객체를 받음
def createTable(create_sql_list , conn):
    # 커서 객체 생성
    # 매개변수로 pymysql.cursors.DictCursor를 적으면 select 쿼리문 했을경우 결과가 딕셔너리 형태로 나옴.
    cursor = conn.cursor(pymysql.cursors.DictCursor)
    for i in create_sql_list :
        #for문을 통하여 list안에 있는 내용을 i를 통해 하나씩 꺼냄.
        #커서를 통해서 create_sql_list안에 있는 내용(쿼리문)을 실행함 
        cursor.execute(i)
```

```python
#데이터 베이스에 Insert 하기 위한 함수
#매개변수로 insert를 하기위해 필요한 정보가 담긴 dictionary, Connection객체를 받음
def insertData(insert_sql_dict, conn):
    cursor = conn.cursor(pymysql.cursors.DictCursor)
    # columns = list(dict.keys())
    # create_sql_dict.keys() -> 데이터를 넣을 테이블명
    
    for i in insert_sql_dict.keys():
        # INSERT 쿼리 작성
        sql = f"INSERT INTO {i}"
        sql = sql + " VALUES "
        # insert_sql_dict[i] -> 안에 딕셔너리가 들어있음
        # insert_sql_dict[i].keys() -> 딕셔너리의 키값들 = 넣고자하는 테이블 컬럼들에 대한 키 값들
        # len(insert_sql_dict[i]['연도']) -> 행의 갯수
        columns = list(insert_sql_dict[i].keys())
        for j in (range(len(insert_sql_dict[i][columns[0]]))):
            # 행의 갯수
            for idx,column in enumerate(columns) :
            # 열의 갯수
                if(idx == 0):
                    # 처음 시작할때는 ( 쓰고시작
                    sql = sql + "("
                sql = sql + f"{insert_sql_dict[i][column][j]}"
                if(idx+1 == len(columns)):
                    sql = sql + ")"
                    # 만약 첫번째 행의 데이터가 다 입력 되었는경우 )로 닫음
                else:
                    sql = sql + ","
                    # 끝나지 않았으면 각 열의 데이터를 구분하기 위해서 , 를 붙여줌
            if(j+1 != len(insert_sql_dict[i][column])):
                #마지막 행인 경우에는 ,를 붙이지 않아야 함
                sql = sql + ','
            # 잘 진행 되는지 확인하기 위한 print문
            print(sql)
        # sql 실행
        cursor.execute(sql)
    # 실행한 명령들을 저장하기 위한 commit 메소드 실행
    conn.commit()
        

    

```

```python
def main():
    # 데이터베이스와 연결
    conn = getConnection() 
    # 테이블 생성
    createTable(create_sql_list , conn)
    # 데이터 삽입
    insertData(insert_sql_dict, conn)
    # 연결 종료
    conn.close()
```

```python
# 메인함수 실행
main()
```
